```python
def pct75(column):
    return column.quantile(0.75)

def pct25(column):
    return column.quantile(0.25)

# sales[['Temperature', 'Weekly_Sales']].agg([pct75, pct25])

def iqr(column):
    return column.quantile(0.75) - column.quantile(0.25)

sales[['Temperature', 'Weekly_Sales']].agg(iqr)
```
